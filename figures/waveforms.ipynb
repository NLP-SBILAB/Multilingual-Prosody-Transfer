```python
# Plot the spectrograms of the audio files
import matplotlib.pyplot as plt
import librosa.display

x1, sr1 = librosa.load(ref_file)
x2, sr2 = librosa.load(syn_file_vc)
x3, sr3 = librosa.load("/home/arnav/BTP/output_speecht5.wav")

import matplotlib.pyplot as plt
import numpy as np

# Assuming x1, x2, x3 are your audio signals and sr1, sr2, sr3 are their respective sample rates

plt.figure(figsize=(14, 8))

# Plot for Reference Audio
plt.subplot(3, 1, 1)
librosa.display.specshow(librosa.amplitude_to_db(np.abs(librosa.stft(x1)), ref=np.max), y_axis='log', x_axis='time')
plt.title('Reference Audio')

# Plot for VC Synthesized Audio
plt.subplot(3, 1, 2)
librosa.display.specshow(librosa.amplitude_to_db(np.abs(librosa.stft(x2)), ref=np.max), y_axis='log', x_axis='time')
plt.title('VC Synthesized Audio')

# Plot for T5 Synthesized Audio
plt.subplot(3, 1, 3)
librosa.display.specshow(librosa.amplitude_to_db(np.abs(librosa.stft(x3)), ref=np.max), y_axis='log', x_axis='time')
plt.title('T5 Synthesized Audio')

plt.tight_layout()
plt.show()
```
